Handling Nan

```python
prime_Data.isnull().sum() #This Shows the sum of the Nan values accross the dataset accornding to its column
```

```python
prime_Data['director'].unique() #we can see the nan value so we can impute it
prime_Data['director'].isnull().sum() #we can see the nan count
```

```python
prime_Data['director'].fillna('Missing',inplace=True) #we can do the Arbitary imputation technique as it is of categorical data
```

```python
prime_Data['director'].unique() #nan is imputed into missing
prime_Data['director'].isnull().sum() #Here we can see the nan count after imputing
```

As of we can impute Arbitary Imputation for cast , country , rating , duration

```python
prime_Data['cast'].fillna('Missing',inplace=True) #imputing cast feature
```

```python
prime_Data.isnull().sum()
```

```python
prime_Data['country'].fillna('Missing',inplace=True) #imputing country feature
```

```python
prime_Data.isnull().sum()
```

```python
prime_Data['rating'].fillna('Missing',inplace=True) #imputing rating feature
```

```python
prime_Data['duration'].fillna('Missing',inplace=True) #imputing duration feature
```

```python
prime_Data.isnull().sum()
```

Format data

```python
prime_Data['date_added'].head()
```

```python
prime_Data['date_added '] = pd.to_datetime(prime_Data['date_added']) #this changes the format of the date_added to datetime datatype
```

```python
prime_Data['date_added '].head()
```

```python
prime_Data['date_added'].isnull().sum()
```

```python
prime_Data.head()
```

```python
prime_Data.drop(columns=['date_added'] ,inplace=True) #we drop the old date added column format
```

```python
prime_Data['date_added '].fillna('ffill',inplace=True) #this is also an imputation where datetime nan is imputed with the previous non-missing value
```

```python
prime_Data['date_added '].isnull().sum() #imputed succesfully
```

```python
prime_Data.isnull().sum()#every nan values are imputed successfully
```

Visualization

```python
plt.figure(figsize=(10,6))
sns.histplot(data=prime_Data ,x='release_year') #release year view
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=prime_Data ,x='type') #comparing Content
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.countplot(data=prime_Data ,x='rating' , order=prime_Data['rating'].value_counts().index) # comparing ratings
plt.show()
```

```python
plt.figure(figsize=(45,20))
sns.countplot(data=prime_Data ,x='country' , order=prime_Data['country'].value_counts().index[:20]) # comparing countries and aquiring top 20 countries
plt.show()
```
