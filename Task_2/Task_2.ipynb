Handling Nan

```python
data.isnull().sum() # finding nan values
```

```python
data['Year'].isnull().sum() # finding nan count in year feature
```

```python
data['Year']=data['Year'].fillna(data['Year'].mode()[0],inplace=False) # filling null values with bfill
```

```python
data['Year'].isnull().sum()# imputed successfully
```

```python
data['Publisher'].isnull().sum() #finding nan count in Publisher feature
```

```python
data['Publisher'].fillna('missing',inplace=True) #Imputing with mode as it fill nan value with frequently occured value
```

```python
data['Publisher'].isnull().sum() #imputed successfully
```

```python
data.isnull().sum()#final check for Nan
```

Changing Datatype

```python
data['North America']=data['North America'].astype(int) #float to int
```

```python
data['Europe']=data['Europe'].astype(int) #float to int
```

```python
data['Japan']=data['Japan'].astype(int) #float to int
```

```python
data['Global']=data['Global'].astype(int) #float to int
```

```python
data['Review']=data['Review'].astype(int) #float to int
```

Visualization

Distribution of game Genres

```python
genre_counts=data['Genre'].value_counts()
plt.figure(figsize=(12,6))
sns.barplot(x=genre_counts.index , y=genre_counts.values) #bargraph
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

Platform Popularity

```python

Platform_counts=data['Platform'].value_counts()
plt.figure(figsize=(12,6))
sns.barplot(x=Platform_counts.index , y=Platform_counts.values) #bargraph
plt.xlabel('Platform')
plt.ylabel('Count')
plt.show()
```

Sales By Region

```python
Region = data[['North America',	'Europe',	'Japan'	,'Rest of World',	'Global']]
```

```python
plt.figure(figsize=(12,6))
Region.sum().plot(kind='bar') #bargraph
plt.xlabel("region")
plt.ylabel("Total sales")
```

total Sales summary

```python
plt.figure(figsize=(12,6))
Region.sum().plot(kind='pie') #piechart
```

Sales over Time

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=data , x='Year',y='Global')
```

year and global sale result

```python
yearsale=data.groupby('Year')['Global'].sum()
```

```python
plt.plot(yearsale.index,yearsale.values,marker='o',linestyle='-')
plt.xlabel('Year')
plt.ylabel('Global sales')
```
